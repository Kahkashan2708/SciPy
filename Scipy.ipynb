```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans2
```

```python
a=np.random.multivariate_normal([0,6],[[2,1],[1,1.5]],size=200)
b=np.random.multivariate_normal([2,0],[[1,-1],[-1,3]],size=400)
c=np.random.multivariate_normal([6,4],[[5,0],[0,1.2]],size=250)
z=np.concatenate((a,b,c))
np.random.shuffle(z)
plt.scatter(z[:,0],z[:,1])
plt.show()
```

```python
k=3
centroids, labels=kmeans2(z,k, minit="points")
print(f"Centroids:{centroids}")
print(f"z={z}")
print(labels)
```

```python
cluster_1=z[labels==0]
cluster_2=z[labels==1]
cluster_3=z[labels==2]

plt.scatter(cluster_1[:,0],cluster_1[:,1],label="Cluster_1")
plt.scatter(cluster_2[:,0],cluster_2[:,1],label="Cluster_2")
plt.scatter(cluster_3[:,0],cluster_3[:,1],label="Cluster_3")
plt.scatter(centroids[:,0],centroids[:,1],label="Centroids",color='#000000')
plt.show()
```

### Interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,11)
y=np.array([2.0,1.9,1.7,1.5,0.5,0.,0.8,2.0,0.9,0.2,2.0])
plt.plot(x,y,"o:")
plt.show()
```

```python
from scipy.interpolate import interp1d
```

```python
predict=interp1d(x,y,kind="quadratic")
# linear, quadratic, cubic, nearest, zero, slinear
x2=np.linspace(0,10,100)
y2=np.array([predict(x) for x in x2])

plt.plot(x,y,"o:")
plt.plot(x2,y2,"ro:")
plt.show()
```

```python
print(f"x: 1.5, y: {predict(1.5)}")
```

```python
predict=interp1d(x,y,kind="nearest")
# linear, quadratic, cubic, nearest, zero, slinear
x2=np.linspace(0,10,100)
y2=np.array([predict(x) for x in x2])

plt.plot(x,y,"o:")
plt.plot(x2,y2,"ro:")
plt.show()
```

# Linear Algebra

```python
import numpy as np
import scipy.linalg as linalg
```

```python
# 3x+4y=10 Linear Equation
# 2x+5y=9

a=np.array([
    [3,4],
    [2,5]])
b=np.array([10,9])
print("Solution: ", linalg.solve(a,b))
```

```python
array=np.array([
           [1,2,3],
           [4,5,6],
           [ 8,8,9]
])
print("Numpy Array:")
print(array)
```

```python
# Find Dterminant
determinant=linalg.det(array)
print("Determinant=",determinant)
```

```python
# Find Inverse
inv_array=linalg.inv(array)
print("Inverse:")
print(inv_array)
```

```python
# Inverse Multiplication
print(array @ inv_array)
```

### Find Eigen value and Eigen vectors

```python
array2=np.array([
          [1,-2,0],
          [-2,1,0],
          [0,0,3]
])
e_value, e_vector=linalg.eig(array2)
print("Eigen Values:",e_value)
print("Eigen Vector:")
print(e_vector)
```

```python
## Singular value Decompositon

p=e_vector
p_inv=linalg.inv(p)
D=[]
for index, e_value in enumerate(e_value):
    row=np.zeros((3))
    row[index]=e_value
    D.append(row)
D=np.array(D)
print(p@D@p_inv)
```

### Integration

```python
import math
from scipy.integrate import quad 
# single integration: quad function
# double integration: dblquad
```

#### Integrate the given function: 


```python
def f(x:float) ->float:
    return math.sin(x)
area, error= quad(f,0,math.pi/2)
print(f"Area: {area}")
print(f"Error: {error}")
```

#### Integrate the given function:


```python
import math
from scipy.integrate import dblquad
```

```python
def f(x,y):
    return x**2+y**2
area, error=dblquad(f,0,1,0,1)
print(f"Area: {area}")
print(f"Error: {error}")
```

```python
def f(x,y,z):
    return x*y*z
result, error= nquad(f,[[0,2],[-2,0],[1,2]]) 
print(f"Integration: {result}")
print(f"Error: {error}")
```

#### maximum and minimum of function (Optimization)

```python
import numpy as np
from scipy.optimize import minimize
```

```python
# Minimize a^2+b^2 with initial condition a=200, b=500

def objective(x):
    a=x[0]
    b=x[1]
    return a**2+b**2
x_0=np.array([200,500])
solution=minimize(objective,x_0,method='SLSQP')
#Minimize a scalar function of one or more variables using Sequential Least Squares Programming (SLSQP).
print(solution)
```

```python
# Minimize a^2+b^2 with initial condition a=700, b=-600 and a+b=100(constraint)

def objective(x):
    a=x[0]
    b=x[1]
    return a**2+b**2

def constraint1(x):
    a=x[0]
    b=x[1]
    return a+b-100
cons=[
    {
        "type": "eq",
        "fun": constraint1
    } ]
x_0=np.array([200,500])
solution=minimize(objective,x_0,method='SLSQP',constraints=cons)
print(solution)
```

```python
# Minimize a^2+b^2 with initial condition a=700, b=-600 with bounds 10<a,b<50
def objective(x):
    a=x[0]
    b=x[1]
    return a**2+b**2
bound=np.array([10,50]) 
bounds=(bound,bound)
x_0=np.array([700,-600])
solution=minimize(objective,x_0,method='SLSQP',bounds=bounds)
print(solution)
```
